# 02 - Data Cleaning

## Objective
Clean the loaded retail dataset by:
- Handling missing values
- Removing duplicates
- Removing cancelled transactions
- Removing invalid quantities and prices
- Creating Revenue feature
- Saving cleaned dataset for further analysis

```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv(
    r"C:\Users\91895\OneDrive\Desktop\NeuralRetail\data\processed\combined_retail.csv"
)

print("Dataset Loaded Successfully")
print(df.shape)

df.head()
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
missing = df.isnull().sum()

missing
```

```python
missing_percent = (missing/len(df))*100

missing_percent.sort_values(ascending=False)
```

```python
df = df.dropna(subset=["Customer ID"])

print(df.shape)
```

```python
print("Before:", df.shape)

df.drop_duplicates(inplace=True)

print("After:", df.shape)
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
df.dtypes
```

```python
df = df[
    ~df["Invoice"].astype(str).str.startswith("C")
]

print(df.shape)
```

```python
df = df[df["Quantity"] > 0]

print(df.shape)
```

```python
df = df[df["Price"] > 0]

print(df.shape)
```

```python
df["Revenue"] = df["Quantity"] * df["Price"]

df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
print(df.shape)
```

```python
os.makedirs(
    r"C:\Users\91895\OneDrive\Desktop\NeuralRetail\data\processed",
    exist_ok=True
)

df.to_csv(
    r"C:\Users\91895\OneDrive\Desktop\NeuralRetail\data\processed\cleaned_retail.csv",
    index=False
)

print("Cleaned Dataset Saved Successfully")
```
