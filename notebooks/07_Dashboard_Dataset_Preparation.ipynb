```python
import pandas as pd
import numpy as np
```

```python
retail = pd.read_csv(r"C:\Users\91895\OneDrive\Desktop\NeuralRetail\data\processed\cleaned_retail.csv")
segments = pd.read_csv(r"C:\Users\91895\OneDrive\Desktop\NeuralRetail\data\processed\customer_segments.csv")
churn = pd.read_csv(r"C:\Users\91895\OneDrive\Desktop\NeuralRetail\data\processed\customer_churn.csv")
forecast = pd.read_csv(r"C:\Users\91895\OneDrive\Desktop\NeuralRetail\data\processed\revenue_forecast.csv")
```

```python
print("Retail Columns:")
print(retail.columns)

print("\nSegments Columns:")
print(segments.columns)

print("\nChurn Columns:")
print(churn.columns)
```

```python
retail['InvoiceDate'] = pd.to_datetime(
    retail['InvoiceDate']
)
```

```python
retail['Year'] = retail['InvoiceDate'].dt.year

retail['Month'] = retail['InvoiceDate'].dt.month

retail['Month_Name'] = retail['InvoiceDate'].dt.month_name()

retail['Quarter'] = retail['InvoiceDate'].dt.quarter

retail['Day'] = retail['InvoiceDate'].dt.day

retail['Weekday'] = retail['InvoiceDate'].dt.day_name()

retail['Weekend_Flag'] = retail['Weekday'].isin(
    ['Saturday', 'Sunday']
)
```

```python
retail['Revenue_Category'] = pd.cut(
    retail['Revenue'],
    bins=[0,50,200,1000,100000],
    labels=[
        'Low',
        'Medium',
        'High',
        'Premium'
    ]
)
```

```python
retail = retail.merge(
    segments[['Customer ID', 'Segment']],
    on='Customer ID',
    how='left'
)
```

```python
retail[['Customer ID','Segment']].head()
```

```python
retail = retail.merge(
    churn[['Customer ID', 'Churn']],
    on='Customer ID',
    how='left'
)
```

```python
retail[['Customer ID','Churn']].head()
```

```python
retail['Churn_Status'] = retail['Churn'].map({
    0: 'Active',
    1: 'At Risk'
})
```

```python
retail['Inventory_Risk'] = np.where(
    retail['Quantity'] < 5,
    'High Risk',
    'Low Risk'
)
```

```python
retail['Order_Size'] = pd.cut(
    retail['Revenue'],
    bins=[0,100,500,1000,100000],
    labels=[
        'Small',
        'Medium',
        'Large',
        'Bulk'
    ]
)
```

```python
dashboard = retail.merge(
    segments[
        [
            "Customer ID",
            "Recency",
            "Frequency",
            "Monetary",
            "Cluster",
            "Segment"
        ]
    ],
    on="Customer ID",
    how="left"
)

dashboard = dashboard.merge(
    churn[
        [
            "Customer ID",
            "Churn",
            "Churn_Prediction"
        ]
    ],
    on="Customer ID",
    how="left"
)
```
